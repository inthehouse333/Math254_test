```r
# Recode the status variableinto 1's and 0's

mydata$cancer_status <- recode(mydata$status, 'M' = 1, 'B' = 0)

table(mydata$status)
table(mydata$cancer_status)
```

```r
mydata %>%
ggplot() +
  aes( x = radius, y = cancer_status) +
  geom_jitter(aes(col = status), height = 0.05, width = 0.01, alpha = 0.5)
```

```r
# Fit the logistic regression model
model <- glm(cancer_status ~ radius, data = mydata, family = 'binomial')

# Add the logistic regression curve to the plot
mydata %>%
  ggplot() +
  aes(x = radius, y = cancer_status) +
  geom_jitter(aes(col = status), height = 0.05, width = 0.01, alpha = 0.5) +
  geom_smooth(method = 'glm', method.args = list(family = 'binomial'), se = FALSE, color = 'red')
```

```r
summary(model)
```

Fitted curve:

$\hat p = \frac{e^{-15.2 + 1.03x}}{1 = e^{-15.2 + 1.03x}}$

where $\hat p$ is the predicted probability from this model.

How to interpret the "slope"?

- First, acknowledge the sign of the "slope" parameter. In this case, the positive sign of the "slope" indicates that probability increase as we move from small to large radius values.

- In order to think about rate of change in probability we need to make some judgements on the x values we use.

```r
mynewdata <- data.frame(radius = c(14, 15, 20, 21))

mypredictions <- predict(model, newdata = mynewdata, type = 'response')

mypredictions
```
